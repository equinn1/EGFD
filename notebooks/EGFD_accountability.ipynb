```python
def get_ffid(platoons,dt,fnm):
    ffid = None
    for platoon in ['A','B','C','D']:
        sc = pyEG.shift_crew(platoons,platoon,dt)
        crew = sc.get_crew()
        for fid in crew.keys():
            try:
                if (crew[fid]['FF'] is not None):
                    if (crew[fid]['FF'].get_name()==fnm):
                        ffid = platoon + fid 
            except KeyError:
                ;
    return(ffid)
```

```python
def findOccurrences(s, ch):
    return [i for i, letter in enumerate(s) if letter == ch]
```

```python
def clean_reason(reason,ffid):
    new_reason = ''
    ix = findOccurrences(reason,'-')
    
    if (ffid is None):
        new_reason += reason
        
    elif ('SWAP-OUT'in reason):
        new_reason += 'SWAP-OUT'
    else:
        if ((',' in reason) & ('-' in reason)):
            ix = findOccurrences(reason,'-')
            ix_last = ix[len(ix)-1]
            new_reason += ffid + reason[ix_last:]

    return(new_reason)        
```

## Build calendar entries for each FF ID and shift

```python
calendar = {}

for ts in shifts.keys():
    calendar[ts] = {}
    calendar[ts]['platoon'] = shifts[ts].get_platoon()
    calendar[ts]['hours'] = shifts[ts].get_hours()
    for ff_name in ffs.keys():
        ff_id = get_ffid(platoons,ts,ff_name)
        if (ff_id is not None):
            calendar[ts][ff_id]= {}
            calendar[ts][ff_id]['ff_name'] = ff_name
            calendar[ts][ff_id]['vacant'] = False
            if (shifts[ts].get_platoon() == ff_id[0:1]):
                calendar[ts][ff_id]['scheduled'] = True
            else:
                calendar[ts][ff_id]['scheduled'] = False
                
    for platoon in platoon_list:                                 #fill in any missing ffids as vacant positions
        for pos in shift_roster:
            ff_id = platoon + pos
            if (ff_id not in calendar[ts].keys()):
                calendar[ts][ff_id] = {}
                calendar[ts][ff_id]['vacant'] = True
                if (shifts[ts].get_platoon() == ff_id[0:1]):
                    calendar[ts][ff_id]['scheduled'] = True
                else:
                    calendar[ts][ff_id]['scheduled'] = False
            
                
for ff_name in acct.keys():
    for ts in acct[ff_name].keys():
        ff_ID = acct[ff_name][ts]['ff_ID']
        try:
            calendar[ts][ff_ID]['type'] = acct[ff_name][ts]['type']
        except KeyError:
            ;
        try:
            calendar[ts][ff_ID]['for_ff'] = acct[ff_name][ts]['for_ff']
        except KeyError:
            ;
        try:
            calendar[ts][ff_ID]['hours'] = acct[ff_name][ts]['hours']
        except KeyError:
            ;
        try:
            calendar[ts][ff_ID]['rank'] = acct[ff_name][ts]['rank']
        except KeyError:
            ;
        try:
            calendar[ts][ff_ID]['reason'] = acct[ff_name][ts]['reason']
        except KeyError:
            ;
        try:
            calendar[ts][ff_ID]['for_ff_id'] = acct[ff_name][ts]['for_ff_id']
        except KeyError:
            ;
        try:
            calendar[ts][ff_ID]['clean_reason'] = acct[ff_name][ts]['clean_reason']
        except KeyError:
            ;
                
```

```python
def who_covered(ts,calendar,ff_id):
    covered_by = None
    for token in calendar[ts].keys():
        try:
            for_ff_id = calendar[ts][token]['for_ff_id']
            if (for_ff_id is not None):
                if (for_ff_id == ff_id):
                    covered_by = token
        except KeyError:
            ;
        except TypeError:
            ;
    return(covered_by)
```

### Build "who covered" fields

```python
for ts in calendar.keys():
    for token in calendar[ts].keys():
        try:
            typ = calendar[ts][token]['type']
            if (typ in ['IOD','VC','SL','SWAP','ADMIN','BER','PRS','CMP-U','BARG']):
                ff_cov = who_covered(ts,calendar,token)
                calendar[ts][token]['covered_by'] = ff_cov
            if (typ == 'OT-VNCY'):
                for tk in calendar[ts].keys():
                    calendar[ts][tk]['covered_by'] = None
                    try:
                        if (isinstance(calendar[ts][tk],dict)):
                            if ((calendar[ts][tk]['scheduled']) & (calendar[ts][tk]['vacant'])):
                                calendar[ts][tk]['covered_by'] = token
                    except AttributeError:
                        ;

        except KeyError:
            ;
        except TypeError:
            ;
        except AttributeError:
            ;
                            
```

### Function returns color code for various shift types

```python
def get_background_color(ts,token):                                   #color codes cells in html table
    bgcolor = '#FFFFFF'
    try:
        cal_ent = calendar[ts][token]
        if ('type' in cal_ent.keys()):
            if (cal_ent['type'] == 'VC'):  bgcolor = '#A0A0A0'        #vacation is gray
            elif (cal_ent['type'] == 'SWAPW'):  bgcolor = '#E8E8E8'   #SWAPW is light gray
            elif (cal_ent['type'] == 'SWAP'):  bgcolor = '#E8E8E8'   #SWAPW is light gray
            elif (cal_ent['type'] == 'SL'):  bgcolor = '#FFFF00'      #sick leave is red
            elif (cal_ent['type'] == 'IOD'):  bgcolor = '#FF00FF'     #IOD is purple
            elif (cal_ent['type'] == 'PRS'):  bgcolor = '#00B2EE'     #personal and bereavement are blue
            elif (cal_ent['type'] == 'BER'):  bgcolor = '#00B2EE'  
            elif (cal_ent['type'] == 'ADMIN'):  bgcolor = '#00B2EE'   #administrative is blue
            elif (cal_ent['type'] == 'OT'):  bgcolor = '#FFC0CB'      #late runs and callbacks are yellow
            elif ('C-' in cal_ent['type']):  bgcolor = '#00FF00'      #collateral and detail are green
            elif ('CS-' in cal_ent['type']):  bgcolor = '#00FF00'
            elif (cal_ent['type'] == 'DP'):  bgcolor = '#00FF00'
            elif ('OT-' in cal_ent['type']):  bgcolor = '#FF0000'
            if (cal_ent['type'] in ['SL','IOD','VC','BARG','PRS','SWAP','ADMIN','CMP-U']):
                if ('covered_by' in cal_ent.keys()):
                    if (cal_ent['covered_by'] is None):
                        bgcolor = '#00FFFF'
                else:
                    bgcolor = '#00FFFF'
                
        bgc = bgcolor
        if ('vacant' in cal_ent.keys()): 
            if ((cal_ent['vacant']) & (cal_ent['scheduled'])): 
                bgcolor = '#00FFFF'
            if (cal_ent['covered_by'] is not None):
                bgcolor = bgc
    except KeyError:
        ;
    except AttributeError:
        ;
    return(bgcolor)
```

### Function builds a row of cells representing a month in the html table

```python
def html_row(platoon,ffn,year,month,hour,calendar,file,last_date):             #function writes a row of a table
        
    ndays = {1:31,2:28,3:31,4:30,5:31,6:30,7:31,8:31,9:30,10:31,11:30,12:31}  #days per month

    token = platoon + ffn                                                     #token is platoon plus FF id
    if (hour == 7):                                                           #day shift row
        full_shift=10
        rowstr = '<tr align="center"><td rowspan="2">' + str(month) + '/' + str(year) + '</td>'
        rowstr +='<td>07:00</td>'
    else:                                                                     #night shift row
        full_shift=14
        rowstr = '<tr><td>17:00</td>'
    
    for d in range(1,ndays[month]+1):                                         #loop through the days of this month
        ts = datetime.datetime(year,month,d,hour,0)                           #timestamp for shift start
        rowstr += '<td bgcolor="' + get_background_color(ts,token) + '">'
        try:                                                                  #shifts are labeled scheduled or not
            scheduled = calendar[ts][token]['scheduled']                      #scheduled show as 'On'
            vacant = calendar[ts][token]['vacant']
            if (scheduled):                                                   #first line
                if (not vacant):
                    rowstr += 'On'
                elif (vacant):
                    rowstr += 'VNCY'
                    if ('covered_by' in calendar[ts][token].keys()):
                        cov_ff = calendar[ts][token]['covered_by']
                        rowstr += '<br>' + calendar[ts][cov_ff]['type']
                if ('type' in calendar[ts][token].keys()):
                    rowstr += '<br><a href="#'  + token + str(ts) + '">'+ calendar[ts][token]['type'] + '</a>'
                else:
                    rowstr += '<br>&nbsp;'
                if ('covered_by' in calendar[ts][token].keys()):
                    cov_ff = calendar[ts][token]['covered_by']
                    if (cov_ff is not None):
                        rowstr += '<br><a href="#'  + cov_ff + str(ts) + '">'+ cov_ff + '</a>'
                else:
                    rowstr += '<br>&nbsp;'
            else:                                                             #others show as blank
                rowstr += '&nbsp'
                if ('type' in calendar[ts][token].keys()):
                    rowstr += '<br><a href="#'  + token + str(ts) + '">'+ calendar[ts][token]['type'] + '</a>'
                if ('for_ff_id' in calendar[ts][token].keys()):
                    if (calendar[ts][token]['for_ff_id'] is not None):
                        for_ff = calendar[ts][token]['for_ff_id']
                        rowstr += '<br><a href="#'  + for_ff + str(ts) + '">'+ for_ff + '</a>'
        except KeyError:
            if (ts.date() < last_date):
                print('key error')
                print(ts)
                print(token)
    rowstr += '</td>'
    rowstr += '</tr>'
    file.write(rowstr)
    return()
```
