## Define scheduled shifts and 8-day cycles

### Function returns ff_id associated with a name at a given point in time

```python
def get_ffid(platoons,dt,fnm):
    ffid = None
    for platoon in platoon_list:
        sc = pyEG.shift_crew(platoons,platoon,dt)
        crew = sc.get_crew()
        for fid in crew.keys():
            try:
                if (crew[fid]['FF'] is not None):
                    if (crew[fid]['FF'].get_name()==fnm):
                        ffid = platoon + fid 
            except KeyError:
                ;
    return(ffid)
```

### Function returns a list with positions of a specific character in a string

```python
def findOccurrences(s, ch):
    return [i for i, letter in enumerate(s) if letter == ch]
```

### Function edits the reason from the accountability log to remove the name

```python
def clean_reason(reason,ffid):
    new_reason = ''
    ix = findOccurrences(reason,'-')
    
    if (ffid is None):
        new_reason += reason
        
    elif ('SWAP-OUT'in reason):
        new_reason += 'SWAP-OUT'
    else:
        if ((',' in reason) & ('-' in reason)):
            ix = findOccurrences(reason,'-')
            ix_last = ix[len(ix)-1]
            new_reason += ffid + reason[ix_last:]

    return(new_reason)        
```

### Function determines whether there is a floater on this platoon at the time of the shift

```python
def have_floater(ts,ff_id):
    have_floater = True
    if ('A' in ff_id):           #Platoon A - has floater prior to 6/1/2017
        if ((ts.year > 2017 ) | ((ts.year == 2017) & (ts.month >= 7))):  have_floater = False
    elif ('B' in ff_id):           #Platoon B - has floater prior to 6/1/2016
        if ((ts.year > 2016 ) | ((ts.year == 2016) & (ts.month >= 7))):  have_floater = False
    elif ('C' in ff_id):           #Platoon C - has floater prior to 6/1/2018
        if ((ts.year > 2018 ) | ((ts.year == 2018) & (ts.month >= 7))):  have_floater = False
    elif ('D' in ff_id):           #Platoon D - has floater prior to 6/1/2018
        if ((ts.year > 2018 ) | ((ts.year == 2018) & (ts.month >= 7))):  have_floater = False
    else: have_floater = None
            
    return(have_floater)
```

### Function returns ff_ID of FF who covered for a specific FF on a specific shift

```python
def who_covered(ts,calendar,ff_id):
    covered_by = None
    for token in calendar[ts].keys():                        #loop through FFs for this shift
        try:
            for_ff_id = calendar[ts][token]['for_ff_id']     #look for ID of FF covered
            if (for_ff_id is not None):                      #if there is one,
                if (for_ff_id == ff_id):                     #see if it is our FF
                    covered_by = token                       #if so set 'covered_by'

        except KeyError:
            ;
        except TypeError:
            ;
        
    return(covered_by)
```

### Function determines hourly wage

```python
def get_wage(ts,ff_id):
    if (ts.date() < datetime.date(2017,7,1)):
        if ((ff_id == 'COF1') | (ff_id == 'AOF3')):              wage = 31.89    #Capt
        elif ('FF' in ff_id):                                    wage = 26.29    #FF
        else:                                                    wage = 28.99    #Lt
    if (ts.date() < datetime.date(2018,7,1)):
        if ((ff_id == 'COF1') | (ff_id == 'AOF3')):              wage = 32.53    #Capt
        elif ('FF' in ff_id):                                    wage = 26.82    #FF
        else:                                                    wage = 29.57    #Lt
            
    return(wage)
```

### Build "who covered" fields in calendar dictionary

```python
for ts in calendar.keys():                                                    #loop through shifts in calendar
    for token in calendar[ts].keys():                                         #keys for this shift
        try:
            typ = calendar[ts][token]['type']                                 #types that require coverage
            if (typ in ['IOD','VC','SL','SWAP','ADMIN','BER','PRS','CMP-U','BARG']):
                ff_cov = who_covered(ts,calendar,token)                       #if required, see who covered
                calendar[ts][token]['covered_by'] = ff_cov
                if (ff_cov is None):
                    if (have_floater(ts,token)):                              #nobody covered, check if floater did
                        calendar[ts][token]['covered_by'] = 'floater'
            elif (typ == 'OT-VNCY'):                                          #if position covered is vacant
                for tokn in calendar[ts].keys():                              # there will be no name of covered
                    if (tokn not in ['platoon','hours']):
                        try:                                                      #if 'scheduled' and 'vacant' mark as
                            if (isinstance(calendar[ts][tokn],dict)):               # covered by this FF
                                if ((calendar[ts][tokn]['scheduled']) & (calendar[ts][tokn]['vacant'])):
                                    calendar[ts][tokn]['covered_by'] = token
                        except AttributeError:
                            ;

        except KeyError:
            ;
        except TypeError:
            ;
        except AttributeError:
            ;
                            
```

### Compute cost for regular scheduled shifts worked

```python
for ts in sorted(calendar.keys()):                         #loop through shifts
    shift_length = calendar[ts]['hours']                   #get length of shift
    for key in sorted(calendar[ts].keys()):                #loop through FFs
        if (('FF' in key) | ('OF' in key)):                #exclude 'platoon' and 'hours' keys
            wage = get_wage(ts,key)                        #get wage
            sched_cost = round(wage*shift_length,2)        #get shift cost at straight time
            if ('type' not in calendar[ts][key].keys()):   #regular shift that was worked or vacant
                if (calendar[ts][key]['scheduled']):
                    calendar[ts][key]['sched_cost'] = sched_cost
            elif ('OT' in calendar[ts][key]['type']):
                acct_hours = calendar[ts][key]['hours']
                calendar[ts][key]['OT_cost'] = 1.5*wage*acct_hours
            elif (('CS-' in calendar[ts][key]['type']) | ('C-' in calendar[ts][key]['type'])):
                acct_hours = calendar[ts][key]['hours']
                if (calendar[ts][key]['type'] == 'C-DD'): calendar[ts][key]['CX_cost'] = acct_hours
                elif (calendar[ts][key]['type'] == 'C-ND'): calendar[ts][key]['CX_cost'] = 20.40*acct_hours
                elif ('C-' in calendar[ts][key]['type']): calendar[ts][key]['CX_cost'] = 24.00*acct_hours
                else: calendar[ts][key]['CX_cost'] = 28.0*acct_hours
            else:
                acct_hours = calendar[ts][key]['hours']
                calendar[ts][key]['sched_cost'] = sched_cost
```

### Function adds hours to dictionary, creating entry if there is none

```python
def add_hours(fd,typ,hours,shift_hours):
    if (('C-' in typ) | ('CS-' in typ) | (typ == 'DP')):    #lump collateral and detail into 'col'
        if ('COL' in fd.keys()):
            fd['COL'] += hours
        else:
            fd['COL'] = hours
    else:                                                   #other types get their own entry
        if (typ in fd.keys()):
            fd[typ] += hours
        else:
            fd[typ] = hours
    if (typ != 'SWAP'):                                     #compute total hours excluding SWAP
        if ('tot' in fd.keys()):
            fd['tot'] += hours
        else:
            fd['tot'] = hours
    if ((typ == 'SWAP') & (hours < shift_hours)):           #handle partial SWAP as some regular, some swap
        if ('reg' in fd.keys()):
            fd['reg'] += shift_hours - hours
        else:
            fd['reg'] = shift_hours - hours
        if ('tot' in fd.keys()):
            fd['tot'] += shift_hours - hours
        else:
            fd['tot'] = shift_hours - hours
    return()
```
