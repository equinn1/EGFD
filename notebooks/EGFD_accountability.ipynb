### Function returns ff_id associated with a name at a given point in time

```python
def get_ffid(platoons,dt,fnm):
    ffid = None
    for platoon in platoon_list:
        sc = pyEG.shift_crew(platoons,platoon,dt)
        crew = sc.get_crew()
        for fid in crew.keys():
            try:
                if (crew[fid]['FF'] is not None):
                    if (crew[fid]['FF'].get_name()==fnm):
                        ffid = platoon + fid 
            except KeyError:
                ;
    return(ffid)
```

## Define scheduled shifts and 8-day cycles

```python
shifts = {}                                                         #dictionary for shifts

cycles = {}                                                         #dictionary for cycles

plist = ['D','C','D','C','A','D','A','D','B','A','B','A','C','B','C','B']   #platoon sequence

ts = datetime.datetime(2017,1,1,7,0,0)                              #earliest shift to generate

maxdate = datetime.date(2018,5,25)                                  #stop when this date is reached

dt = ts.date()                                                      #date part of shift start

ix = 0                                                              #index into platoon array

while(dt < maxdate):                                                 #continue until end date is reached
    platoon = plist[ix]                                             #get platoon
    shifts[ts] = pyEG.shift(ts,platoon,platoons)                    #create shift object
    if (ix in [0,4,8,12]):                                          #create cycle object if start of cycle
        cycles[ts] = {}
        cycles[ts]['platoon'] = platoon
    if (ts.hour == 7):         ts += datetime.timedelta(hours=10)   #increment datetime by hours in shift
    else:         ts += datetime.timedelta(hours=14)
    dt = ts.date()                                                  #new date part of shift start
    ix = (ix+1)%16                                                  #new platoon index
    
for ts in cycles.keys():
    cycles[ts]['cycle'] = pyEG.cycle(ts,shifts)
    
    

print('shifts')
print(len(shifts))
print('cycles')
print(len(cycles))
```

### Function returns a list with positions of a specific character in a string

```python
def findOccurrences(s, ch):
    return [i for i, letter in enumerate(s) if letter == ch]
```

### Function edits the reason from the accountability log to remove the name

```python
def clean_reason(reason,ffid):
    new_reason = ''
    ix = findOccurrences(reason,'-')
    
    if (ffid is None):
        new_reason += reason
        
    elif ('SWAP-OUT'in reason):
        new_reason += 'SWAP-OUT'
    else:
        if ((',' in reason) & ('-' in reason)):
            ix = findOccurrences(reason,'-')
            ix_last = ix[len(ix)-1]
            new_reason += ffid + reason[ix_last:]

    return(new_reason)        
```

### Function determines whether there is a floater on this platoon at the time of the shift

```python
def have_floater(ts,ff_id):
    have_floater = True
    if ('A' in ff_id):           #Platoon A - has floater prior to 6/1/2017
        if ((ts.year > 2017 ) | ((ts.year == 2017) & (ts.month >= 7))):  have_floater = False
    elif ('B' in ff_id):           #Platoon B - has floater prior to 6/1/2016
        if ((ts.year > 2016 ) | ((ts.year == 2016) & (ts.month >= 7))):  have_floater = False
    elif ('C' in ff_id):           #Platoon C - has floater prior to 6/1/2018
        if ((ts.year > 2018 ) | ((ts.year == 2018) & (ts.month >= 7))):  have_floater = False
    elif ('D' in ff_id):           #Platoon D - has floater prior to 6/1/2018
        if ((ts.year > 2018 ) | ((ts.year == 2018) & (ts.month >= 7))):  have_floater = False
    else: have_floater = None
            
    return(have_floater)
```

### Function returns wage cost of shift by type

```python
def get_cost(ts,rank,hours,ty):
    hr = get_wage(ts,rank)
    if ('OT' in ty):
        cost = hr*hours*1.5
    elif (ty in ['C-DD','C-ND']):                             #dispatch
        if ((ts.hour >= 7) & (ts.hour <= 17)):
            cost = 22.44*hours                                 #day rate is $22.44
        else:
            cost = 20.40*hours                                 #night rate is $20.40
    elif('CS-' in ty):
        cost = 28.00*hours                                     #collateral duty supervisor rate is $28.00
    elif('C-' in ty):
        cost = 24.00*hours                                     #collateral duty rate is $20.00
    elif(ty in ['DP','DFP']):
        cost = 40.0*hours                                      #detail rate is $35 (town) $40 (private) assume private
    elif(ty == 'SWAP'):
        cost = wage*hours                                      #SWAP is paid regular wage
    elif(ty =='SWAPW'):
        cost = 0.0                                             #SWAP-OUT is paid $0
    elif(ty =='CMP-E'):
        cost = 0.0                                             #CMP-E is paid $0
    elif(ty =='CMP-U'):
        cost = 0.0                                             #CMP-U is paid $0
    elif(ty in ['VC','SL','IOD','PRS','BER','ADMIN','ACT','BARG','SCH','EMER']):
        cost = wage*hours                                      #various reasons not worked
    else:
        cost = None
        
    return(cost)
```

### Function builds a row of cells representing a month in the html table

```python
def html_row(platoon,ffn,year,month,hour,calendar,file,last_date):             #function writes a row of a table
        
    ndays = {1:31,2:28,3:31,4:30,5:31,6:30,7:31,8:31,9:30,10:31,11:30,12:31}  #days per month

    token = platoon + ffn                                                     #token is platoon plus FF id
    if (hour == 7):                                                           #day shift row
        full_shift=10
        rowstr = '<tr align="center"><td rowspan="2">' + str(month) + '/' + str(year) + '</td>'
        rowstr +='<td>07:00</td>'
    else:                                                                     #night shift row
        full_shift=14
        rowstr = '<tr><td>17:00</td>'
    
    for d in range(1,ndays[month]+1):                                         #loop through the days of this month
        ts = datetime.datetime(year,month,d,hour,0)                           #timestamp for shift start
        rowstr += '<td bgcolor="' + get_background_color(ts,token,calendar) + '">'
        covered_vacancy = False
        try:                                                                  
            scheduled = calendar[ts][token]['scheduled']                      #boolean - is shift scheduled?
            vacant = calendar[ts][token]['vacant']                            #boolean - is position vacant?
            if (scheduled):                                                   #do scheduled shift
                if (not vacant):                                              #show 'On' unless vacant
                    rowstr += 'On Duty'
                elif (vacant):                                                #scheduled and vacant
                    rowstr += 'On Duty<br>VNCY'
                    if ('covered_by' in calendar[ts][token].keys()):
                        cov_ff = calendar[ts][token]['covered_by']
                        covered_vacancy = True
                        rowstr += '<br>' + calendar[ts][cov_ff]['type'] + '<br>' + cov_ff
                if ('type' in calendar[ts][token].keys()):
                    rowstr += '<br><a href="#'  + token + str(ts) + '">'+ calendar[ts][token]['type'] + '</a>'
                    if ('covered_by' in calendar[ts][token].keys()):
                        cov_ff = calendar[ts][token]['covered_by']
                        if (cov_ff is not None):
                            rowstr += '<br><a href="#'  + cov_ff + str(ts) + '">'+ cov_ff + '</a>'
                try:
                    if (('sched_cost' in calendar[ts][token].keys()) &  (not vacant)):
                        rowstr += '<br>' + '${:,.2f}'.format(calendar[ts][token]['sched_cost'])
                    else:
                        rowstr += '<br>&nbsp;'
                except AttributeError:
                    rowstr += '<br>&nbsp;'

                rowstr += '<br>&nbsp;'

            else:                                                             #others show as blank
                rowstr += '&nbsp'
                if ('type' in calendar[ts][token].keys()):
                    rowstr += '<br><a href="#'  + token + str(ts) + '">'+ calendar[ts][token]['type'] + '</a>'
                if ('for_ff_id' in calendar[ts][token].keys()):
                    if (calendar[ts][token]['for_ff_id'] is not None):
                        for_ff = calendar[ts][token]['for_ff_id']
                        rowstr += '<br><a href="#'  + for_ff + str(ts) + '">'+ for_ff + '</a>'
                if ('CX_cost' in calendar[ts][token].keys()):
                        rowstr += '<br>' + '${:,.2f}'.format(calendar[ts][token]['CX_cost'])
                if ('OT_cost' in calendar[ts][token].keys()):
                        rowstr += '<br>' + '${:,.2f}'.format(calendar[ts][token]['OT_cost'])
        except KeyError:
            if (ts.date() < last_date):
                print('key error')
                print(ts)
                print(token)
    rowstr += '</td>'
    rowstr += '</tr>'
    file.write(rowstr)
    return()
```

## Pay Period Analysis

```python
pay_period = {}

dt19 = datetime.datetime(2018,7,30,7,0,0)

for dt in sorted(calendar.keys()):
    if (dt > datetime.datetime(2017,6,25,6,0,0)):
        shift_hours = calendar[dt]['hours']
        td = dt - datetime.datetime(2017,6,25,7,0,0)
        seq_pay_period2 = 24*td.days + td.seconds//3600 
        payroll_week = 1+ int(seq_pay_period2/(24*7))
        pay_period_start = datetime.datetime(2017,6,25,7,0,0) + datetime.timedelta(days=7*(payroll_week-1))
        ppsd = pay_period_start.date()
        for ffid in calendar[dt].keys():
            if (('OF' in ffid) | ('FF' in ffid)):
                wage = get_wage(dt,ffid)
                wage19 = get_wage(dt19,ffid)
                if (ffid not in pay_period.keys()):  pay_period[ffid] = {}
                if (ppsd not in pay_period[ffid].keys()): 
                    pay_period[ffid][ppsd] = {}
                    pay_period[ffid][ppsd]['wage'] = wage
                    pay_period[ffid][ppsd]['wage19'] = wage19
                    pay_period[ffid][ppsd]['scheduled'] = 0.0
                    pay_period[ffid][ppsd]['not_worked'] = 0.0
                    pay_period[ffid][ppsd]['vacation'] = 0.0
                    pay_period[ffid][ppsd]['sick_leave'] = 0.0
                    pay_period[ffid][ppsd]['IOD'] = 0.0
                    pay_period[ffid][ppsd]['other'] = 0.0
                    pay_period[ffid][ppsd]['SWAP'] = 0.0
                    pay_period[ffid][ppsd]['SWAPW'] = 0.0
                    pay_period[ffid][ppsd]['OT'] = 0.0
                    pay_period[ffid][ppsd]['collateral'] = 0.0
                    pay_period[ffid][ppsd]['regular_wages'] = 0.0
                    pay_period[ffid][ppsd]['regular_wages19'] = 0.0
                    pay_period[ffid][ppsd]['overtime_wages'] = 0.0
                    pay_period[ffid][ppsd]['overtime_wages19'] = 0.0
                    pay_period[ffid][ppsd]['overtime_adjustment'] = 0.0
                    pay_period[ffid][ppsd]['overtime_adjustment2'] = 0.0
                    pay_period[ffid][ppsd]['CX_cost'] = 0.0
                    pay_period[ffid][ppsd]['total_wages'] = 0.0
                if (calendar[dt][ffid]['scheduled']):
                    pay_period[ffid][ppsd]['scheduled']+=shift_hours
                    pay_period[ffid][ppsd]['regular_wages']+=shift_hours*wage
                    pay_period[ffid][ppsd]['regular_wages19']+=shift_hours*wage19
                    if ('type' in calendar[dt][ffid].keys()):
                        hours = calendar[dt][ffid]['hours']
                        typ   = calendar[dt][ffid]['type']
                        if (typ in ['VC','SL','IOD','PRS','BER','BARG','ADMIN','SWAP','ACT','SCH','CMP-U','EMER','ACTOT']):
                            pay_period[ffid][ppsd]['not_worked'] += hours
                            if (typ=='VC'): 
                                pay_period[ffid][ppsd]['vacation']+=hours
                            elif (typ == 'SL'):
                                pay_period[ffid][ppsd]['sick_leave']+=hours
                            elif (typ == 'IOD'):
                                pay_period[ffid][ppsd]['IOD']+=hours
                            elif (typ == 'SWAP'):
                                pay_period[ffid][ppsd]['SWAP']+=hours
                            else:
                                pay_period[ffid][ppsd]['other']+=hours
                        else:
                            print('type mismatch with scheduled ' + typ + ' ' + ffid + ' ' + str(dt))    
                else:
                    if ('type' in calendar[dt][ffid].keys()):
                        typ = calendar[dt][ffid]['type']
                        hours = calendar[dt][ffid]['hours']
                        if ('OT' in typ):
                            pay_period[ffid][ppsd]['OT']+=hours
                            pay_period[ffid][ppsd]['overtime_wages']+=1.5*hours*wage
                            pay_period[ffid][ppsd]['overtime_wages19']+=1.5*hours*wage19
                            
                        elif (('C-' in typ) | ('CS-' in typ) | (typ == 'DP')):
                            hours = calendar[dt][ffid]['hours']
                            CX_cost = calendar[dt][ffid]['CX_cost']
                            pay_period[ffid][ppsd]['collateral']+=hours
                            pay_period[ffid][ppsd]['CX_cost']+=CX_cost
                        elif (typ == 'SWAPW'):
                            pay_period[ffid][ppsd]['SWAPW']+=hours
                            
                pay_period[ffid][ppsd]['not_worked'] = max(0, \
                            pay_period[ffid][ppsd]['not_worked']-pay_period[ffid][ppsd]['SWAPW'])
                if ((pay_period[ffid][ppsd]['not_worked']>0.0) & (pay_period[ffid][ppsd]['OT']>0.0)):
                    pay_period[ffid][ppsd]['overtime_adjustment'] = -0.5*wage*min(pay_period[ffid][ppsd]['not_worked'], \
                                    pay_period[ffid][ppsd]['OT'])
                    not_worked2 = max(0,pay_period[ffid][ppsd]['not_worked']-pay_period[ffid][ppsd]['collateral'])
                    pay_period[ffid][ppsd]['overtime_adjustment2'] = -0.5*wage*min(not_worked2,pay_period[ffid][ppsd]['OT'])
                pay_period[ffid][ppsd]['total_wages'] = pay_period[ffid][ppsd]['CX_cost'] + \
                        pay_period[ffid][ppsd]['overtime_wages']+pay_period[ffid][ppsd]['regular_wages']
               
```
