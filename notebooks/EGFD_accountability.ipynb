## Define scheduled shifts and 8-day cycles

```python
shifts = {}                                                         #dictionary for shifts

cycles = {}                                                         #dictionary for cycles

plist = ['D','C','D','C','A','D','A','D','B','A','B','A','C','B','C','B']   #platoon sequence

ts = datetime.datetime(2017,1,1,7,0,0)                              #earliest shift to generate

maxdate = datetime.date(2018,5,25)                                  #stop when this date is reached

dt = ts.date()                                                      #date part of shift start

ix = 0                                                              #index into platoon array

while(dt < maxdate):                                                 #continue until end date is reached
    platoon = plist[ix]                                             #get platoon
    shifts[ts] = pyEG.shift(ts,platoon,platoons)                    #create shift object
    if (ix in [0,4,8,12]):                                          #create cycle object if start of cycle
        cycles[ts] = {}
        cycles[ts]['platoon'] = platoon
    if (ts.hour == 7):         ts += datetime.timedelta(hours=10)   #increment datetime by hours in shift
    else:         ts += datetime.timedelta(hours=14)
    dt = ts.date()                                                  #new date part of shift start
    ix = (ix+1)%16                                                  #new platoon index
    
for ts in cycles.keys():
    cycles[ts]['cycle'] = pyEG.cycle(ts,shifts)
    
    

print('shifts')
print(len(shifts))
print('cycles')
print(len(cycles))
```

### Function returns ff_id associated with a name at a given point in time

```python
def get_ffid(platoons,dt,fnm):
    ffid = None
    for platoon in platoon_list:
        sc = pyEG.shift_crew(platoons,platoon,dt)
        crew = sc.get_crew()
        for fid in crew.keys():
            try:
                if (crew[fid]['FF'] is not None):
                    if (crew[fid]['FF'].get_name()==fnm):
                        ffid = platoon + fid 
            except KeyError:
                ;
    return(ffid)
```

### Function returns a list with positions of a specific character in a string

```python
def findOccurrences(s, ch):
    return [i for i, letter in enumerate(s) if letter == ch]
```

### Function edits the reason from the accountability log to remove the name

```python
def clean_reason(reason,ffid):
    new_reason = ''
    ix = findOccurrences(reason,'-')
    
    if (ffid is None):
        new_reason += reason
        
    elif ('SWAP-OUT'in reason):
        new_reason += 'SWAP-OUT'
    else:
        if ((',' in reason) & ('-' in reason)):
            ix = findOccurrences(reason,'-')
            ix_last = ix[len(ix)-1]
            new_reason += ffid + reason[ix_last:]

    return(new_reason)        
```

## Build calendar entries for each FF ID and shift

```python
calendar = {}                                                   #calendar dictionary

for ts in shifts.keys():                                        #loop through shifts
    calendar[ts] = {}                                           #calendar dictionary entry for this shift
    calendar[ts]['platoon'] = shifts[ts].get_platoon()          #platoon on duty
    calendar[ts]['hours'] = shifts[ts].get_hours()              #hours in this shift
    for ff_name in ffs.keys():                                  #loop through FF names
        ff_id = get_ffid(platoons,ts,ff_name)                   #get ff_id assigned for this shift, if any
        if (ff_id is not None):                                 #if name is assigned for this shift
            calendar[ts][ff_id]= {}                             #entry for ff_id
            calendar[ts][ff_id]['ff_name'] = ff_name            #name
            calendar[ts][ff_id]['vacant'] = False               #position is not vacant on this shift
            if (shifts[ts].get_platoon() == ff_id[0:1]):        #for the platoon on duty
                calendar[ts][ff_id]['scheduled'] = True         # flag scheduled=True
            else:                                               #otherwise flag scheduled=False
                calendar[ts][ff_id]['scheduled'] = False
                
    for platoon in platoon_list:                                 #fill in any missing ffids as vacant positions
        for pos in shift_roster:                                 #loop through platoons and shift roster
            ff_id = platoon + pos                                #token is platoon+position in roster
            if (ff_id not in calendar[ts].keys()):               #if we haven't created an entry for this pos
                calendar[ts][ff_id] = {}                         #create one now with 'vacant'=True
                calendar[ts][ff_id]['vacant'] = True             #
                if (shifts[ts].get_platoon() == ff_id[0:1]):     #if platoon is on duty set 'scheduled'=True
                    calendar[ts][ff_id]['scheduled'] = True
                else:                                            #otherwise 'scheduled' = False
                    calendar[ts][ff_id]['scheduled'] = False
            
                
for ff_name in acct.keys():                                              #fill in data from accountability logs
    for ts in acct[ff_name].keys():                                      #loop through entries for this FF
        ff_ID = acct[ff_name][ts]['ff_ID']                               #get ff_ID: platoon+pos
        try:
            calendar[ts][ff_ID]['type'] = acct[ff_name][ts]['type']      #copy 'type' to calendar
        except KeyError:
            ;
        try:
            calendar[ts][ff_ID]['for_ff'] = acct[ff_name][ts]['for_ff']  #copy FF covered to calendar
        except KeyError:
            ;
        try:
            calendar[ts][ff_ID]['hours'] = acct[ff_name][ts]['hours']    #copy number of hours to calendar
        except KeyError:
            ;
        try:
            calendar[ts][ff_ID]['rank'] = acct[ff_name][ts]['rank']      #copy rank for this shift to calendar
        except KeyError:
            ;
        try:
            calendar[ts][ff_ID]['reason'] = acct[ff_name][ts]['reason']  #copy reason to calendar
        except KeyError:
            ;
        try:
            calendar[ts][ff_ID]['for_ff_id'] = acct[ff_name][ts]['for_ff_id']  #copy ff_ID of covered FF
        except KeyError:
            ;
        try:
            calendar[ts][ff_ID]['clean_reason'] = acct[ff_name][ts]['clean_reason']  #copy reason with name removed
        except KeyError:
            ;
                
```

### Function returns ff_ID of FF who covered for a specific FF on a specific shift

```python
def who_covered(ts,calendar,ff_id):
    covered_by = None
    for token in calendar[ts].keys():                        #loop through FFs for this shift
        try:
#            if ((ts.year==2017) & (ts.month==4) & (ts.day==5)):
#                print(ts)
#                print(ff_id + ' ' + token)
#                print(calendar[ts][token])
            for_ff_id = calendar[ts][token]['for_ff_id']     #look for ID of FF covered
            if (for_ff_id is not None):                      #if there is one,
                if (for_ff_id == ff_id):                     #see if it is our FF
                    covered_by = token                       #if so set 'covered_by'
        except KeyError:
            ;
        except TypeError:
            ;
    return(covered_by)
```

### Build "who covered" fields in calendar dictionary

```python
for ts in calendar.keys():                                                    #loop through shifts in calendar
    for token in calendar[ts].keys():                                         #keys for this shift
        try:
            typ = calendar[ts][token]['type']                                 #types that require coverage
            if (typ in ['IOD','VC','SL','SWAP','ADMIN','BER','PRS','CMP-U','BARG']):
                ff_cov = who_covered(ts,calendar,token)                       #if required, see who covered
                calendar[ts][token]['covered_by'] = ff_cov
#                if ((ts.year==2017) & (ts.month==4) & (ts.day==5)):
#                    print(str(ts.hour) + ' ' +token + ' ' + ff_cov)
#                    print(calendar[ts][token])
            elif (typ == 'OT-VNCY'):                                          #if position covered is vacant
                for tokn in calendar[ts].keys():                              # there will be no name of covered
                    if (tokn not in ['platoon','hours']):
                        #calendar[ts][tokn]['covered_by'] = None                     #see which FFs are not covered
                        try:                                                      #if 'scheduled' and 'vacant' mark as
                            if (isinstance(calendar[ts][tokn],dict)):               # covered by this FF
                                if ((calendar[ts][tokn]['scheduled']) & (calendar[ts][tokn]['vacant'])):
                                    calendar[ts][tokn]['covered_by'] = token
                        except AttributeError:
                            ;

        except KeyError:
            ;
        except TypeError:
            ;
        except AttributeError:
            ;
                            
```

### Function returns color code for various shift types

```python
def get_background_color(ts,token):                                   #color codes cells in html table
    bgcolor = '#FFFFFF'
    try:
        cal_ent = calendar[ts][token]
        if ('type' in cal_ent.keys()):
            if (cal_ent['type'] == 'VC'):  bgcolor = '#A0A0A0'        #vacation is gray
            elif (cal_ent['type'] == 'SWAPW'):  bgcolor = '#E8E8E8'   #SWAPW is light gray
            elif (cal_ent['type'] == 'SWAP'):  bgcolor = '#E8E8E8'   #SWAPW is light gray
            elif (cal_ent['type'] == 'SL'):  bgcolor = '#FFFF00'      #sick leave is red
            elif (cal_ent['type'] == 'IOD'):  bgcolor = '#FF00FF'     #IOD is purple
            elif (cal_ent['type'] == 'PRS'):  bgcolor = '#00B2EE'     #personal and bereavement are blue
            elif (cal_ent['type'] == 'BER'):  bgcolor = '#00B2EE'  
            elif (cal_ent['type'] == 'ADMIN'):  bgcolor = '#00B2EE'   #administrative is blue
            elif (cal_ent['type'] == 'OT'):  bgcolor = '#FFC0CB'      #late runs and callbacks are pink
            elif ('C-' in cal_ent['type']):  bgcolor = '#00FF00'      #collateral and detail are green
            elif ('CS-' in cal_ent['type']):  bgcolor = '#00FF00'
            elif (cal_ent['type'] == 'DP'):  bgcolor = '#00FF00'
            elif ('OT-' in cal_ent['type']):  bgcolor = '#FF0000'
            if (cal_ent['type'] in ['SL','IOD','VC','BARG','PRS','SWAP','ADMIN','CMP-U','VNCY']):
                if ('covered_by' in cal_ent.keys()):
                    if (cal_ent['covered_by'] is None):
                        bgcolor = '#FF8C00'
                    else:
                        bgcolor = '#FF0000'
                else:
                    bgcolor = '#FF8C00'
                
        bgc = bgcolor
        if ('vacant' in cal_ent.keys()): 
            if ((cal_ent['vacant']) & (cal_ent['scheduled'])): 
                bgcolor = '#FF8C00'
            if (cal_ent['covered_by'] is not None):
                bgcolor = '#FF0000'   #bgc
    except KeyError:
        ;
    except AttributeError:
        ;
    return(bgcolor)
```

### Function builds a row of cells representing a month in the html table

```python
def html_row(platoon,ffn,year,month,hour,calendar,file,last_date):             #function writes a row of a table
        
    ndays = {1:31,2:28,3:31,4:30,5:31,6:30,7:31,8:31,9:30,10:31,11:30,12:31}  #days per month

    token = platoon + ffn                                                     #token is platoon plus FF id
    if (hour == 7):                                                           #day shift row
        full_shift=10
        rowstr = '<tr align="center"><td rowspan="2">' + str(month) + '/' + str(year) + '</td>'
        rowstr +='<td>07:00</td>'
    else:                                                                     #night shift row
        full_shift=14
        rowstr = '<tr><td>17:00</td>'
    
    for d in range(1,ndays[month]+1):                                         #loop through the days of this month
        ts = datetime.datetime(year,month,d,hour,0)                           #timestamp for shift start
        rowstr += '<td bgcolor="' + get_background_color(ts,token) + '">'
        try:                                                                  #shifts are labeled scheduled or not
            scheduled = calendar[ts][token]['scheduled']                      #scheduled show as 'On'
            vacant = calendar[ts][token]['vacant']
            if (scheduled):                                                   #first line
                if (not vacant):
                    rowstr += 'On'
                elif (vacant):
                    rowstr += 'VNCY'
                    if ('covered_by' in calendar[ts][token].keys()):
                        cov_ff = calendar[ts][token]['covered_by']
                        rowstr += '<br>' + calendar[ts][cov_ff]['type']
                if ('type' in calendar[ts][token].keys()):
                    rowstr += '<br><a href="#'  + token + str(ts) + '">'+ calendar[ts][token]['type'] + '</a>'
                else:
                    rowstr += '<br>&nbsp;'
                if ('covered_by' in calendar[ts][token].keys()):
                    cov_ff = calendar[ts][token]['covered_by']
                    if (cov_ff is not None):
                        rowstr += '<br><a href="#'  + cov_ff + str(ts) + '">'+ cov_ff + '</a>'
                else:
                    rowstr += '<br>&nbsp;'
            else:                                                             #others show as blank
                rowstr += '&nbsp'
                if ('type' in calendar[ts][token].keys()):
                    rowstr += '<br><a href="#'  + token + str(ts) + '">'+ calendar[ts][token]['type'] + '</a>'
                if ('for_ff_id' in calendar[ts][token].keys()):
                    if (calendar[ts][token]['for_ff_id'] is not None):
                        for_ff = calendar[ts][token]['for_ff_id']
                        rowstr += '<br><a href="#'  + for_ff + str(ts) + '">'+ for_ff + '</a>'
        except KeyError:
            if (ts.date() < last_date):
                print('key error')
                print(ts)
                print(token)
    rowstr += '</td>'
    rowstr += '</tr>'
    file.write(rowstr)
    return()
```

### Function adds hours to dictionary, creating entry if there is none

```python
def add_hours(fd,typ,hours,shift_hours):
    if (('C-' in typ) | ('CS-' in typ) | (typ == 'DP')):    #lump collateral and detail into 'col'
        if ('COL' in fd.keys()):
            fd['COL'] += hours
        else:
            fd['COL'] = hours
    else:                                                   #other types get their own entry
        if (typ in fd.keys()):
            fd[typ] += hours
        else:
            fd[typ] = hours
    if (typ != 'SWAP'):                                     #compute total hours excluding SWAP
        if ('tot' in fd.keys()):
            fd['tot'] += hours
        else:
            fd['tot'] = hours
    if ((typ == 'SWAP') & (hours < shift_hours)):           #handle partial SWAP as some regular, some swap
        if ('reg' in fd.keys()):
            fd['reg'] += shift_hours - hours
        else:
            fd['reg'] = shift_hours - hours
        if ('tot' in fd.keys()):
            fd['tot'] += shift_hours - hours
        else:
            fd['tot'] = shift_hours - hours
    return()
```
